## Analysis overview

As a parent of a child under 10, I'm subjected to a barrage of different cartoons on a daily basis - the current favourite in our house is ["Hot Wheels Let's Race"](https://www.netflix.com/title/81566729). First released on Netflix in early 2024, the series follows "six young hopefuls at the Hot Wheels Ultimate Garage Racing Camp as they learn the skills they need to become the next generation of racers". The show is set within a universe centered around the Hot Wheels toy line by Mattel and the cars used in the series are modeled on real-world Hot Wheels toys, as is also the case with other related Hot Wheels media.
<br>

I recently became aware of 2 video games that are part of the Hot Wheels media franchise (["Hot Wheels Unleashed"](https://en.wikipedia.org/wiki/Hot_Wheels_Unleashed) and ["Hot Wheels Unleashed 2: Turbocharged"](https://en.wikipedia.org/wiki/Hot_Wheels_Unleashed_2:_Turbocharged)) and I asked myself the following questions;

* How many cars are present across all three media items, and which media contains the highest count of models?
* Were any of the cars in the TV series present in one/both of the games, and if so how many?
* Which of the games features more of the cars from the TV series on its roster?
* Which game is the most cost-effective purchase for racing as a car from the TV series?

* From a list of top-5 favourite car models from the TV series, which game is the best purchase to be able to race as a favourite car?

## Data cleaning

The resulting DataFrames are concatenated into a single DataFrame per media item, and cleaned and formatted with the following actions to facilitate data analysis:

* Renaming the ```Model Name``` column to ```model_name```
* Stripping whitespace
* Dropping any duplicates from ```model_name```
* Adding a one-hot endcode of a model's prescence within the DataFrame (i.e. ```in_tv_show``` = 1, ```in_game_1``` = 1, ```in_game_2``` = 1)
* Setting all values to lowercase
* Removing non-alpha characters
* Sorting alphabetically by ```model_name```
<br>

Final DataFrames for each media item can be saved as a local .CSV file using the corresponding commented statements, and loaded using the commented statements in code cell 8.
<br>
<br>


````python
# Cleaning tv_show columns and content
tv_show.rename(columns={'Model Name': 'model_name'},inplace=True)
tv_show.columns = tv_show.columns.str.strip()
tv_show = tv_show.drop_duplicates(subset=["model_name"])
tv_show.insert(1, "in_tv_show", 1)
tv_show.loc[:, "model_name"] = tv_show.loc[:, "model_name"].str.lower()
tv_show.loc[:, "model_name"] = tv_show.loc[:, "model_name"].str.replace("-", " ")
tv_show.loc[:, "model_name"] = tv_show.loc[:, "model_name"].str.replace("'", " ")
tv_show.loc[:, "model_name"] = tv_show.loc[:, "model_name"].str.strip()
tv_show = tv_show.sort_values(by=["model_name"]).reset_index()

#tv_show.to_csv('tv_show.csv')
tv_show
````

````python
# Concatenating and cleaning game_1 columns and content
game_1 = pd.concat([game_1_base, game_1_trackbuilder, game_1_free_single, game_1_free_thanksgiving, game_1_free_christmas, game_1_free_funpack, game_1_beefed_up, game_1_sportscars, game_1_street_beasts, game_1_seasonpass_1_single, game_1_seasonpass_1_batman, game_1_seasonpass_2_single, game_1_seasonpass_2_monster, game_1_seasonpass_3_single, game_1_seasonpass_3_looney], ignore_index=True)
game_1.rename(columns={'Model Name': 'model_name'},inplace=True)
game_1.columns = game_1.columns.str.strip()
game_1 = game_1.drop_duplicates(subset=["model_name"])
game_1.insert(2, "in_game_1", 1)
game_1.loc[:, "model_name"] = game_1.loc[:, "model_name"].str.lower()
game_1.loc[:, "model_name"] = game_1.loc[:, "model_name"].str.replace("-", " ")
game_1.loc[:, "model_name"] = game_1.loc[:, "model_name"].str.replace("'", " ")
game_1.loc[:, "model_name"] = game_1.loc[:, "model_name"].str.replace(".", " ")
game_1.loc[:, "model_name"] = game_1.loc[:, "model_name"].str.strip()
game_1 = game_1.sort_values(by=["model_name"]).reset_index()

#game_1.to_csv('game_1.csv')
game_1
````

````python
# Concatenating and cleaning game_2 columns and content
game_2 = pd.concat([game_2_base, game_2_trackbuilder, game_2_free_manga, game_2_free_highway, game_2_free_accel_1, game_2_free_megabites, game_2_free_accel_2, game_2_free_accel_3, game_2_seasonpass_1_single, game_2_seasonpass_1_accel_allstar, game_2_seasonpass_1_speedandstyle, game_2_seasonpass_1_fastx, game_2_seasonpass_1_accel_xpan, game_2_seasonpass_1_honda, game_2_seasonpass_1_italy, game_2_seasonpass_2_highway, game_2_seasonpass_2_monster, game_2_seasonpass_2_faf, game_2_seasonpass_2_oldgold, game_2_seasonpass_2_alien, game_2_seasonpass_2_merc], ignore_index=True)
game_2.rename(columns={'Model Name': 'model_name'},inplace=True)
game_2.columns = game_2.columns.str.strip()
game_2 = game_2.drop_duplicates(subset=["model_name"])
game_2.insert(2, "in_game_2", 1)
game_2.loc[:, "model_name"] = game_2.loc[:, "model_name"].str.lower()
game_2.loc[:, "model_name"] = game_2.loc[:, "model_name"].str.replace(r"(returning)", "")
game_2.loc[:, "model_name"] = game_2.loc[:, "model_name"].str.replace("-", " ")
game_2.loc[:, "model_name"] = game_2.loc[:, "model_name"].str.replace("'", " ")
game_2.loc[:, "model_name"] = game_2.loc[:, "model_name"].str.replace(".", " ")
game_2.loc[:, "model_name"] = game_2.loc[:, "model_name"].str.strip()
game_2 = game_2.sort_values(by=["model_name"]).reset_index()

#game_2.to_csv('game_2.csv')
game_2
````

The cleaned DataFrames are now combined using left merge on the tv_show dataset, keeping all models present in the TV series. NUll values in the ```in_game_1``` and ```in_game_2``` one-hot encode columns are filled with ```0``` to preserve the datatype as ```int```.
<br>
<br>

A filtered DataFrame is then created, keeping only the ```model_name``` and one-hot encode columns.
<br>
<br>


````python
# Creating left-merged DataFrame, keeping all from tv_show
combined = pd.merge(tv_show, game_1, on="model_name", how="left").merge(game_2, on="model_name", how="left")
combined[['in_tv_show', 'in_game_1', 'in_game_2']] = combined[['in_tv_show', 'in_game_1', 'in_game_2']].fillna(0).astype(int)

final_combined = combined[["model_name", "in_tv_show", "in_game_1", "in_game_2"]]
final_combined
````

The following pie charts can now be created using matplotlib to visualise the data:

1. The presence of the TV series car models in either/both Game 1 and Game 2
2. The proportion of Game 1 car models present in the TV series
3. The proportion of Game 2 car models present in the TV series
<br>
<br>

It's evident that of the 104 car models referenced in the TV series, most are not present in either game. 26 TV series models are present in both games, with a further 8 models unique to Game 2 and 5 models unique to Game 1. This indicates that **Game 2 has the higher presence of TV series car models in its roster**. However, while Game 2 boasts a higher overall count of TV series models, **Game 1 has a higher proportion of its roster present in the TV series** at 23% compared to Game 2's 18%.
<br>
<br>


````python
# Pie charts
plt.pie(final_combined[['in_tv_show', 'in_game_1', 'in_game_2']].value_counts(), labels=final_combined[['in_tv_show', 'in_game_1', 'in_game_2']].value_counts())
plt.legend(["Not in games", "In series, both games", "In series, game 2", "In series, game 1"], loc=1)
plt.title('1. Presence of TV series roster in both games')
plt.show()
plt.clf()
````

````python
# Creating left-merged DataFrame, keeping all from in_game_1
left_in_game_1 = pd.merge(game_1, tv_show, on="model_name", how="left")
left_in_game_1[['in_tv_show', 'in_game_1']] = left_in_game_1[['in_tv_show', 'in_game_1']].fillna(0).astype(int)

# Pie chart 2
plt.pie(left_in_game_1[['in_tv_show']].value_counts(), labels=left_in_game_1[['in_tv_show']].value_counts(), autopct='%.0f%%')
plt.legend(["Not in TV series", "In TV series"], loc=1)
plt.title('2. Proportion of Game 1 roster present in the TV series')
plt.show()
plt.clf()
````

````python
# Creating left-merged DataFrame, keeping all from in_game_2
left_in_game_2 = pd.merge(game_2, tv_show, on="model_name", how="left")
left_in_game_2[['in_tv_show', 'in_game_2']] = left_in_game_2[['in_tv_show', 'in_game_2']].fillna(0).astype(int)

# Pie chart 2
plt.pie(left_in_game_2[['in_tv_show']].value_counts(), labels=left_in_game_2[['in_tv_show']].value_counts(), autopct='%.0f%%')
plt.title('3. Proportion of Game 2 roster present in TV series')
plt.legend(["Not present", "Present"], loc='best')
plt.show()
plt.clf()
````

A bar chart can also be created using seaborn to visualise the volume of models present across the different media - we can see 31 models present in Game 1 and 34 models present in Game 2, reaffirming that Game 2 contains more models present in the TV series.
<br>
<br>


````python
# Count of '1' values in each column
count_in_tv_show = final_combined['in_tv_show'].sum()
count_in_game_1 = final_combined['in_game_1'].sum()
count_in_game_2 = final_combined['in_game_2'].sum()

# Bar chart
plot_data = pd.DataFrame({
    'Category': ['TV Show', 'Game 1', 'Game 2'],
    'Count': [count_in_tv_show, count_in_game_1, count_in_game_2]
})

plt.figure(figsize=(10, 6))
barplot = sns.barplot(data=plot_data, x='Category', y='Count', palette="muted", hue='Category')

for index, row in plot_data.iterrows():
    barplot.text(index, row.Count + 0.5, f'{row.Count}', color='black', ha="center")

plt.title('Count of TV series roster present in Games relative to TV series')
plt.ylabel('Count')
plt.xlabel('Media item')
plt.ylim(0, 120)
plt.grid(axis="y")
plt.show()
plt.clf()
````

At the time of data analysis (17/07/2024), digital versions of both games are being listed at a discounted price on the Playstation Store.

The current discounted prices for the digital versions of these games are:

* Game 1: £10.49
* Game 2: £18.74
<br>

At discounted prices, it's evident that **Game 1 is the most cost-effective purchase for racing both as a car from the TV series and as a car from the overall roster volume**, being 21p cheaper-per-car for the former and 2p cheaper-per-car for the latter.
<br>
<br>

The usual retail prices for the digital versions of the games are:

* Game 1: £69.99
* Game 2: £74.99
<br>

At recommended-retail-prices, it's evident that **Game 2 is the most cost-effective purchase for both racing as a car from the TV series and as a car from the overall roster volume**, being 5p cheaper-per-car for the former and 14p cheaper-per-car for the latter.
<br>
<br>


````python
# Setting count and price variables
game_1_tv_count = 31
game_1_count = 133
game_2_tv_count = 34
game_2_count = 191

game_1_disc = 1049
game_2_disc = 1874
game_1_rrp = 6999
game_2_rrp = 7499

# Calculating cost-per-car (cpc) for both sets of counts and prices for both games
cpc_game_1_tv_disc = game_1_disc / game_1_tv_count
cpc_game_1_disc = game_1_disc / game_1_count
cpc_game_2_tv_disc = game_2_disc / game_2_tv_count
cpc_game_2_disc = game_2_disc / game_2_count

cpc_game_1_tv_rrp = game_1_rrp / game_1_tv_count
cpc_game_1_rrp = game_1_rrp / game_1_count
cpc_game_2_tv_rrp = game_2_rrp / game_2_tv_count
cpc_game_2_rrp = game_2_rrp / game_2_count

# Printing results
print("Cost-per-tv-show-car for Game 1 at the discounted price: £{}".format(round(cpc_game_1_tv_disc / 100, 2)))
print("Cost-per-car for Game 1 at the discounted price: £{}\n".format(round(cpc_game_1_disc / 100, 2)))
print("Cost-per-tv-show-car for Game 2 at the discounted price: £{}".format(round(cpc_game_2_tv_disc / 100, 2)))
print("Cost-per-car for Game 2 at the discounted price: £{}\n".format(round(cpc_game_2_disc / 100, 2)))

print("Cost-per-tv-show-car for Game 1 at the retail price: £{}".format(round(cpc_game_1_tv_rrp / 100, 2)))
print("Cost-per-car for Game 1 at the retail price: £{}\n".format(round(cpc_game_1_rrp / 100, 2)))
print("Cost-per-tv-show-car for Game 2 at the retail price: £{}".format(round(cpc_game_2_tv_rrp / 100, 2)))
print("Cost-per-car for Game 2 at the retail price: £{}\n".format(round(cpc_game_2_rrp / 100, 2)))
````

To further analyse the favourite list, the DataFrame can be reshaped to long format using ```melt``` and a crosstab created to be able to view stacked bars visualising the specific models from the favourite list present in both games.
<br>
<br>

It's evident that the same 3 car models from the favourite list are present across both games, meaning that **the top-5 favourite car models should have no direct impact on game purchase choice**.
<br>
<br>


````python
# Reshape the DataFrame to long format
fav_long = favourite_frame.melt(id_vars='model_name', var_name='category', value_name='presence')

# Filter to keep only the rows where presence is 1
fav_long = fav_long[fav_long['presence'] == 1]

# Create a crosstab to count occurrences
fav_crosstab = pd.crosstab(fav_long['category'], fav_long['model_name'])

# Plotting
fav_crosstab.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.title('Count of favourite car models in Games relative to TV series')
plt.xlabel('Media item')
plt.ylabel('Count')
plt.legend(title='Car Model')
plt.show()
plt.clf()
````
